```python
import pandas as pd

# Creating employee dataset
data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Save to CSV file
df.to_csv("employees.csv", index=False)

print("employees.csv file created successfully!")
import pandas as pd
# Load the dataset
df = pd.read_csv("employees.csv")
# (a) Shape of the DataFrame
print("Shape of the DataFrame (rows, columns):", df.shape)
# (b) Summary of the DataFrame
print("\nSummary of the DataFrame:")
print(df.info())
# (c) Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())
# (d) First 5 rows
print("\nFirst 5 rows of the dataset:")
print(df.head())
#  Last 3 rows
print("\nLast 3 rows of the dataset:")
print(df.tail(3))

# (e) Calculate statistics
average_salary = df["Salary"].mean()
total_bonus = df["Bonus"].sum()
youngest_age = df["Age"].min()
highest_rating = df["Rating"].max()

print("Average Salary:", average_salary)
print("Total Bonus Paid:", total_bonus)
print("Youngest Employee Age:", youngest_age)
print("Highest Performance Rating:", highest_rating)

# (f) Sort DataFrame by Salary in descending order
df_sorted = df.sort_values(by="Salary", ascending=False)
print("\nDataFrame sorted by Salary in descending order:")
print(df_sorted)

# (g) Categorize employees based on performance rating
def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df["Performance_Category"] = df["Rating"].apply(categorize_performance)
print("\nAdded Performance_Category column:")
print(df[["Name", "Rating", "Performance_Category"]])

# (h) Identify missing values
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)

# (i) Rename Employee_ID column to ID
df.rename(columns={"Employee_ID": "ID"}, inplace=True)
print("\nRenamed Employee_ID to ID:")
print(df.head())

# (j) Find employees with more than 5 years of experience
experienced_employees = df[df["Years_of_Experience"] > 5]
print("\nEmployees with more than 5 years of experience:")
print(experienced_employees)

# Find employees who belong to the IT department
it_employees = df[df["Department"] == "IT"]
print("\nEmployees in the IT department:")
print(it_employees)

# (k) Add a new column 'Tax' which deducts 10% of Salary
df["Tax"] = df["Salary"] * 0.10
print("\nAdded Tax column (10% deduction from Salary):")
print(df[["Name", "Salary", "Tax"]])

# (l) Save the modified DataFrame to a new CSV file
df.to_csv("employees_modified.csv", index=False)
print("\nModified dataset saved as 'employees_modified.csv'")



```
